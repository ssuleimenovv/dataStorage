```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# some imports
```

```python
# 1. Launch data
def load_data(file_path):
    df = pd.read_csv(file_path)
    return df

file_path = "appearances.csv"
df = load_data(file_path)
```

```python
# 2. General information about data
def data_overview(df):
    print("\ngeneral info:")
    df.info()
    display(df.head())

data_overview(df)
```

```python
# 3. Checking for missing values:
def check_missing_values(df):
    missing_values = df.isnull().sum()
    print("\nMissing values in dataset:")
    print(missing_values)

check_missing_values(df)
```

```python
# 4. Descriptive statistics
def descriptive_statistics(df):
    print("Descriptive statistics:")
    display(df.describe())

descriptive_statistics(df)
```

```python
# 5. Top-5 players in goals:
def top_scorers(df):
    print("\nTop-5 players in goals:")
    display(df[['playerID', 'goals']].sort_values(by='goals', ascending=False).head(5))

top_scorers(df)
```

```python
# 6. Visualization of data
def plot_histogram(df, column, bins=30):
    plt.figure(figsize=(10, 5))
    sns.histplot(df[column], bins=bins, kde=True)
    plt.title(f"Distribution {column}")
    plt.title(column)
    plt.title("Frequency")
    plt.show()

plot_histogram(df, 'time')
```

```python
# 7. Boxplot of 
def plot_boxplot(df, column):
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=df[column])
    plt.title(f"Distribution {column}")
    plt.show()

plot_boxplot(df, 'goals')
```

```python
# 8. Correlation matrix
def plot_correlation_matrix(df):
    plt.figure(figsize=(12, 6))
    numeric_df = df.select_dtypes(include=['number'])  # remains integer values
    sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
    plt.title("Correlation matrix")
    plt.show()

plot_correlation_matrix(df)
```

```python
# 9. Business Recommendations
def business_recommendations(df):
    print("\nBusiness Recommendations:")
    
    # 1. Optimizing playing time
    if 'time' in df.columns:
        avg_time = df['time'].mean()
        print(f"Average playing time: {avg_time:.2f} minutes. Consider adjusting player rotation strategies.")
    
    # 2. Analyzing the effectiveness of forwards
    if 'goals' in df.columns:
        top_scorers = df[df['goals'] > df['goals'].quantile(0.9)]
        print(f"Top 10% players by goals:\n{top_scorers[['playerID', 'goals']].to_string(index=False)}")
    
    # 3. Identifying key success factors
    numeric_df = df.select_dtypes(include=['number'])
    if not numeric_df.empty:
        print("\nFactors that have the greatest impact on team success:")
        corr_values = numeric_df.head()['goals'].sort_values(ascending=False)
        print(corr_values)
    
    # 4. Using data for future predictions
    print("\nIt is recommended to use data for predictive modeling to forecast team success based on player characteristics.")

business_recommendations(df)
```

```python
# 10. Additional Queries
def additional_queries(df):
    print("\nAdditional Queries:")

    # 1. Average number of goals per year
    if 'year' in df.columns and 'goals' in df.columns:
        avg_goals_per_year = df.groupby('year')['goals'].mean()
        print("Average number of goals per year:")
        print(avg_goals_per_year)

    # 2. Number of matches per country
    if 'country' in df.columns:
        matches_by_country = df['country'].value_counts()
        print("Number of matches per country:")
        print(matches_by_country)

    # 3. Average age of players
    if 'age' in df.columns:
        avg_age = df['age'].mean()
        print(f"Average age of players: {avg_age:.2f} years")

    # 4. Players with the most matches
    if 'playerID' in df.columns and 'matches' in df.columns:
        top_players_by_matches = df.groupby('player')['matches'].sum().sort_values(ascending=False).head(10)
        print("Top 10 players with the most matches:")
        print(top_players_by_matches)

    # 5. Distribution of goals by position
    if 'position' in df.columns and 'goals' in df.columns:
        goals_by_position = df.groupby('position')['goals'].sum()
        print("Distribution of goals by position:")
        print(goals_by_position)

additional_queries(df)
```
